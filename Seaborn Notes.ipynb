## Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA) is an approach to analyzing data sets to summarize their main characteristics, often using statistical graphics and other data visualization methods. The primary goal of EDA is to see what the data can tell us beyond the formal modeling or hypothesis testing task.

**Uses:**
- **Understanding Data:** EDA helps in understanding the underlying structure of the data.
- **Identifying Patterns:** It assists in identifying patterns, anomalies, and trends within the data.
- **Detecting Outliers:** It helps in detecting outliers and errors in the data.
- **Hypothesis Formation:** EDA aids in formulating hypotheses for further analysis.
- **Data Cleaning:** It can be used to identify missing values, duplicates, and other data issues that need to be addressed.
- **Feature Selection:** EDA helps in selecting the most important features for modeling.



## Seaborn

Seaborn is a Python data visualization library based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

**Uses:**
- **Statistical Plots:** Seaborn is used for creating a variety of statistical plots such as bar plots, violin plots, box plots, and scatter plots.
- **Complex Visualizations:** It simplifies the process of creating complex visualizations like heatmaps, pair plots, and facet grids.
- **Aesthetic Customization:** Seaborn comes with built-in themes and color palettes to make visualizations more attractive and easier to interpret.

**Benefits in EDA:**
- **Simplifies Plotting:** Seaborn simplifies the creation of complex plots, making it easier to explore and understand the data.
- **Integration with Pandas:** Seaborn works well with Pandas DataFrames, making it easier to visualize data directly from data frames.
- **Statistical Insights:** The library provides tools for visualizing the distribution of data and relationships between variables, which are essential for EDA.
- **Customization:** It allows for easy customization of plots, which can help in highlighting specific insights and findings during EDA.
  
Using Seaborn during EDA can enhance the effectiveness of the analysis by providing clear, concise, and visually appealing graphics that can reveal underlying patterns and insights in the data.

### Box Plot
A box plot helps to maintain the distribution of quantitative data in such a way that it facilitates the comparisons between variables or across levels of a categorical variable. The main body of the box plot showing the quartiles and the median’s confidence intervals if enabled. The medians have horizontal lines at the median of each box and while whiskers have the vertical lines extending to the most extreme, non-outlier data points and caps are the horizontal lines at the ends of the whiskers.

### Swarm Plot
Swarm plots are similar to strip plots, however they add spread to the data points. Similar to strip plots, they allow you to better understand the distribution of data along concentrated areas. ### Q-Q plot
When the quantiles of two variables are plotted against each other, then the plot obtained is known as quantile – quantile plot or qqplot. This plot provides a summary of whether the distributions of two variables are similar or not with respect to the locations.

```python
sm.qqplot(df['total_bill'])
py.show()
```

```python
sm.qqplot(df['total_bill'],line='45')
py.show()
```

### Hexbin Plot
Hexbin plots take in lists of X and Y values (and an optional list of Z values) and returns what looks somewhat similar to a scatter plot, but where the entire graphing space has been divided into hexagons (like a honeycomb) and all points have been grouped into their respective hexagonal regions with a color gradient indicating the density (or mean Z values) of each hexagonal area.

```python
sns.jointplot(x=df['total_bill'],y=df['tip'],kind='hex')
```

## Mosaic Plot
Mosaic Plot is a graphical method for visualizing data from two or more Qualitative data. It’s also called crosstabs or two-way table and it’s used to summarize the relationship among several categorical variables. The Mosaic Plot is based on conditional probabilities.

```python
data = {'John': 7, 'Joe': 10, 'James': 5, 'Kate': 3}
mosaic(data, title='Basic mosaic plot')
plt.show()
```
